```python
#RUTAS
menu = "/content/drive/MyDrive/proyecto_shaker_chaos/1.png"
sene1 = "/content/drive/MyDrive/proyecto_shaker_chaos/2.png"
scene2 = "/content/drive/MyDrive/proyecto_shaker_chaos/3.png"
kitchen_background = "/content/drive/MyDrive/proyecto_shaker_chaos/kitchen.png"

#bebidas
milk_scene_img = "/content/drive/MyDrive/proyecto_shaker_chaos/5.png"
water_scene_img = "/content/drive/MyDrive/proyecto_shaker_chaos/6.png"
beer_scene_img = "/content/drive/MyDrive/proyecto_shaker_chaos/7.png"
#frutas
milk_strawberry_img = "/content/drive/MyDrive/proyecto_shaker_chaos/8.png" # leche + fresa
milk_lemon_img = "/content/drive/MyDrive/proyecto_shaker_chaos/9.png"      # leche + limón
milk_banana_img = "/content/drive/MyDrive/proyecto_shaker_chaos/10.png"     # leche + banana
water_strawberry_img = "/content/drive/MyDrive/proyecto_shaker_chaos/11.png" # agua + fresa
water_lemon_img = "/content/drive/MyDrive/proyecto_shaker_chaos/12.png"      # agua + limón
water_banana_img = "/content/drive/MyDrive/proyecto_shaker_chaos/13.png"     # agua + bananah
beer_strawberry_img = "/content/drive/MyDrive/proyecto_shaker_chaos/14.png"  # cerveza + fresa
beer_lemon_img = "/content/drive/MyDrive/proyecto_shaker_chaos/15.png"       # cerveza + limón
beer_banana_img = "/content/drive/MyDrive/proyecto_shaker_chaos/16.png"      # cerveza + banana
#extras
milk_strawberry_toy_img = "/content/drive/MyDrive/proyecto_shaker_chaos/17.png" # Leche + Fresa + Juguete
milk_strawberry_leaf_img = "/content/drive/MyDrive/proyecto_shaker_chaos/18.png" # Leche + Fresa + Hoja
milk_lemon_toy_img = "/content/drive/MyDrive/proyecto_shaker_chaos/19.png"      # Leche + Limón + Juguete
milk_lemon_leaf_img = "/content/drive/MyDrive/proyecto_shaker_chaos/20.png"      # Leche + Limón + Hoja
milk_banana_toy_img = "/content/drive/MyDrive/proyecto_shaker_chaos/21.png"     # Leche + Banana + Juguete
milk_banana_leaf_img = "/content/drive/MyDrive/proyecto_shaker_chaos/22.png"     # Leche + Banana + Hoja
water_strawberry_toy_img = "/content/drive/MyDrive/proyecto_shaker_chaos/23.png" # Agua + Fresa + Juguete
water_strawberry_leaf_img = "/content/drive/MyDrive/proyecto_shaker_chaos/24.png" # Agua + Fresa + Hoja
water_lemon_toy_img = "/content/drive/MyDrive/proyecto_shaker_chaos/25.png"      # Agua + Limón + Juguete
water_lemon_leaf_img = "/content/drive/MyDrive/proyecto_shaker_chaos/26.png"      # Agua + Limón + Hoja
water_banana_toy_img = "/content/drive/MyDrive/proyecto_shaker_chaos/27.png"     # Agua + Banana + Juguete
water_banana_leaf_img = "/content/drive/MyDrive/proyecto_shaker_chaos/28.png"     # Agua + Banana + Hoja
beer_strawberry_toy_img = "/content/drive/MyDrive/proyecto_shaker_chaos/29.png"  # Cerveza + Fresa + Juguete
beer_strawberry_leaf_img = "/content/drive/MyDrive/proyecto_shaker_chaos/30.png"  # Cerveza + Fresa + Hoja
beer_lemon_toy_img = "/content/drive/MyDrive/proyecto_shaker_chaos/31.png"       # Cerveza + Limón + Juguete
beer_lemon_leaf_img = "/content/drive/MyDrive/proyecto_shaker_chaos/32.png"       # Cerveza + Limón + Hoja
beer_banana_toy_img = "/content/drive/MyDrive/proyecto_shaker_chaos/33.png"      # Cerveza + Banana + Juguete
beer_banana_leaf_img = "/content/drive/MyDrive/proyecto_shaker_chaos/34.png"      # Cerveza + Banana + Hoja
#después de beber
boom_img = "/content/drive/MyDrive/proyecto_shaker_chaos/boom.png"  # exploto la cocina
drink_shaker_chaos_img = "/content/drive/MyDrive/proyecto_shaker_chaos/36.png"  # beber shaker chaos
#audios
boom_audio = "/content/drive/MyDrive/proyecto_shaker_chaos/audios/boom.mp3"
buho_audio = "/content/drive/MyDrive/proyecto_shaker_chaos/audios/buho.mp3"
caballo_audio= "/content/drive/MyDrive/proyecto_shaker_chaos/audios/caballo.mp3"
cerdo_audio= "/content/drive/MyDrive/proyecto_shaker_chaos/audios/cerdo.mp3"
colibri_audio="/content/drive/MyDrive/proyecto_shaker_chaos/audios/colibri.mp3"
elefante_audio="/content/drive/MyDrive/proyecto_shaker_chaos/audios/elefante.mp3"
gato_audio="/content/drive/MyDrive/proyecto_shaker_chaos/audios/gato.mp3"
leon_audio="/content/drive/MyDrive/proyecto_shaker_chaos/audios/leon.mp3"
mono_audio="/content/drive/MyDrive/proyecto_shaker_chaos/audios/mono.mp3"
oveja_audio="/content/drive/MyDrive/proyecto_shaker_chaos/audios/oveja.mp3"
perro_audio="/content/drive/MyDrive/proyecto_shaker_chaos/audios/perro.mp3"
rana_audio="/content/drive/MyDrive/proyecto_shaker_chaos/audios/rana.mp3"
vaca_audio="/content/drive/MyDrive/proyecto_shaker_chaos/audios/vaca.mp3"

def show_menu():
    clear_output()

    print("MENÚ PRINCIPAL")
    display(IPython.display.Image(filename=menu, width=700))

    start_btn = widgets.Button(description="START", button_style='success')
    exit_btn = widgets.Button(description="EXIT", button_style='danger')

    display(start_btn, exit_btn)

    start_btn.on_click(lambda b: show_scene1())
    exit_btn.on_click(lambda b: clear_output())


# --- ESCENA 1 ---
def show_scene1():
    clear_output()

    print("ESCENA 1")
    display(IPython.display.Image(filename=sene1, width=700))

    next_btn = widgets.Button(description="NEXT →", button_style='primary')
    display(next_btn)

    next_btn.on_click(lambda b: show_scene2())


# --- ESCENA 2 ---
def show_scene2():
    clear_output()

    print("ESCENA 2")
    display(IPython.display.Image(filename=scene2, width=700))

    next_btn = widgets.Button(description="NEXT TO KITCHEN", button_style='info')
    display(next_btn)

    next_btn.on_click(lambda b: show_kitchen_scene())

# --- ESCENA BEBER SHAKER CHAOS ---
def show_shaker_chaos_scene():
    global selected_beverage, selected_fruit, selected_item # Access global selected_item
    clear_output()

    final_audio = None
    final_image = None
    result_message = "¡Shaker Chaos Completado!"

    # Logic for combinations
    if selected_beverage == "leche":
        if selected_fruit == "fresa":
            if selected_item == "juguete":
                final_audio = boom_audio
                final_image = boom_img
                result_message = "¡BOOM! Algo salió mal..."
            elif selected_item == "hoja":
                final_audio = caballo_audio
                final_image = drink_shaker_chaos_img
        elif selected_fruit == "limón":
            if selected_item == "juguete":
                final_audio = cerdo_audio
                final_image = drink_shaker_chaos_img
            elif selected_item == "hoja":
                final_audio = colibri_audio
                final_image = drink_shaker_chaos_img
        elif selected_fruit == "banana":
            if selected_item == "juguete":
                final_audio = elefante_audio
                final_image = drink_shaker_chaos_img
            elif selected_item == "hoja":
                final_audio = gato_audio
                final_image = drink_shaker_chaos_img
    elif selected_beverage == "agua":
        if selected_fruit == "fresa":
            if selected_item == "juguete":
                final_audio = leon_audio
                final_image = drink_shaker_chaos_img
            elif selected_item == "hoja":
                final_audio = mono_audio
                final_image = drink_shaker_chaos_img
        elif selected_fruit == "limón":
            if selected_item == "juguete":
                final_audio = oveja_audio
                final_image = drink_shaker_chaos_img
            elif selected_item == "hoja":
                final_audio = perro_audio
                final_image = drink_shaker_chaos_img
        elif selected_fruit == "banana":
            if selected_item == "juguete":
                final_audio = boom_audio
                final_image = boom_img
                result_message = "¡BOOM! Algo salió mal..."
            elif selected_item == "hoja":
                final_audio = rana_audio
                final_image = drink_shaker_chaos_img
    elif selected_beverage == "cerveza":
        if selected_fruit == "fresa":
            if selected_item == "juguete":
                final_audio = vaca_audio
                final_image = drink_shaker_chaos_img
            elif selected_item == "hoja":
                final_audio = buho_audio
                final_image = drink_shaker_chaos_img
        elif selected_fruit == "limón":
            if selected_item == "juguete":
                final_audio = caballo_audio
                final_image = drink_shaker_chaos_img
            elif selected_item == "hoja":
                final_audio = cerdo_audio
                final_image = drink_shaker_chaos_img
        elif selected_fruit == "banana":
            if selected_item == "juguete":
                final_audio = colibri_audio # Corrected from colibrí_audio to colibri_audio
                final_image = drink_shaker_chaos_img
            elif selected_item == "hoja":
                final_audio = boom_audio
                final_image = boom_img
                result_message = "¡BOOM! Algo salió mal..."

    # Play selected audio and display image
    if final_audio and final_image:
        print(result_message)
        display(IPython.display.Audio(final_audio, autoplay=True))
        display(IPython.display.Image(filename=final_image, width=700))
    else:
        # Fallback if no specific combination is matched
        print("¡BOOM! Algo salió mal... (Combinación no reconocida)")
        display(IPython.display.Audio(boom_audio, autoplay=True))
        display(IPython.display.Image(filename=boom_img, width=700))

    back_to_menu_btn = widgets.Button(description="VOLVER AL MENÚ PRINCIPAL", button_style='primary')
    display(back_to_menu_btn)
    back_to_menu_btn.on_click(lambda b: show_menu())


# --- ESCENA COCINA CON SELECCIÓN DE BEBIDA Y FRUTA Y OBJETO ---
def show_kitchen_scene():
    global selected_beverage, selected_fruit, selected_item
    clear_output()

    print("COCINA - SELECCIONA TU BEBIDA")
    display(IPython.display.Image(filename=kitchen_background, width=700))

    milk_btn = widgets.Button(description="LECHE", button_style='info')
    water_btn = widgets.Button(description="AGUA", button_style='info')
    beer_btn = widgets.Button(description="CERVEZA", button_style='info')
    confirm_beverage_btn = widgets.Button(description="Confirmar Selección de Bebida", button_style='primary')

    output_message_beverage = widgets.Output()

    def on_milk_click(b):
        global selected_beverage
        selected_beverage = "leche"
        with output_message_beverage:
            clear_output()
            print(f"Has seleccionado: {selected_beverage.upper()}")

    def on_water_click(b):
        global selected_beverage
        selected_beverage = "agua"
        with output_message_beverage:
            clear_output()
            print(f"Has seleccionado: {selected_beverage.upper()}")

    def on_beer_click(b):
        global selected_beverage
        selected_beverage = "cerveza"
        with output_message_beverage:
            clear_output()
            print(f"Has seleccionado: {selected_beverage.upper()}")

    def show_fruit_selection_on_beverage_screen():
        clear_output()
        current_beverage_img = None
        if selected_beverage == "leche":
            print("¡Aquí tienes tu leche!")
            current_beverage_img = milk_scene_img
        elif selected_beverage == "agua":
            print("¡Aquí tienes tu agua!")
            current_beverage_img = water_scene_img
        elif selected_beverage == "cerveza":
            print("¡Aquí tienes tu cerveza!")
            current_beverage_img = beer_scene_img
        else:
            print("No has seleccionado ninguna bebida. Por favor, elige una.")
            # If no beverage selected, return to kitchen scene to force selection
            display(IPython.display.Image(filename=kitchen_background, width=700))
            display(widgets.HBox([milk_btn, water_btn, beer_btn]), confirm_beverage_btn, output_message_beverage)
            return

        if current_beverage_img:
            display(IPython.display.Image(filename=current_beverage_img, width=700))
            print("\nAhora, SELECCIONA TU FRUTA:")

            # Fruit selection buttons
            strawberry_btn = widgets.Button(description="FRESA", button_style='info')
            lemon_btn = widgets.Button(description="LIMÓN", button_style='info')
            banana_btn = widgets.Button(description="BANANA", button_style='info')
            confirm_fruit_btn = widgets.Button(description="Confirmar Selección de Fruta", button_style='success')

            output_message_fruit = widgets.Output()

            def on_strawberry_click_fruit(b):
                global selected_fruit
                selected_fruit = "fresa"
                with output_message_fruit:
                    clear_output()
                    print(f"Has seleccionado: {selected_fruit.upper()}")

            def on_lemon_click_fruit(b):
                global selected_fruit
                selected_fruit = "limón"
                with output_message_fruit:
                    clear_output()
                    print(f"Has seleccionado: {selected_fruit.upper()}")

            def on_banana_click_fruit(b):
                global selected_fruit
                selected_fruit = "banana"
                with output_message_fruit:
                    clear_output()
                    print(f"Has seleccionado: {selected_fruit.upper()}")

            def show_item_selection_on_fruit_screen():
                clear_output()
                # Re-display beverage and fruit image if available
                current_combination_img = None
                if selected_beverage == "leche":
                    if selected_fruit == "fresa": current_combination_img = milk_strawberry_img
                    elif selected_fruit == "limón": current_combination_img = milk_lemon_img
                    elif selected_fruit == "banana": current_combination_img = milk_banana_img
                elif selected_beverage == "agua":
                    if selected_fruit == "fresa": current_combination_img = water_strawberry_img
                    elif selected_fruit == "limón": current_combination_img = water_lemon_img
                    elif selected_fruit == "banana": current_combination_img = water_banana_img
                elif selected_beverage == "cerveza":
                    if selected_fruit == "fresa": current_combination_img = beer_strawberry_img
                    elif selected_fruit == "limón": current_combination_img = beer_lemon_img
                    elif selected_fruit == "banana": current_combination_img = beer_banana_img

                if current_combination_img:
                    display(IPython.display.Image(filename=current_combination_img, width=700))
                elif selected_fruit:
                    # Fallback to just fruit image if combination not found
                    print(f"Has seleccionado {selected_beverage.upper()} con {selected_fruit.upper()}.")
                else:
                    print("No has seleccionado ninguna fruta. Por favor, elige una.")
                    # If no fruit selected, return to fruit selection to force selection
                    display(widgets.VBox([
                        widgets.HBox([strawberry_btn, lemon_btn, banana_btn]),
                        confirm_fruit_btn,
                        output_message_fruit
                    ]))
                    return

                print("\nFinalmente, SELECCIONA UN OBJETO:")
                toy_btn = widgets.Button(description="JUGUETE", button_style='info')
                leaf_btn = widgets.Button(description="HOJA", button_style='info')
                confirm_item_btn = widgets.Button(description="Confirmar Selección de Objeto", button_style='success')

                output_message_item = widgets.Output()

                def on_toy_click_item(b):
                    global selected_item
                    selected_item = "juguete"
                    with output_message_item:
                        clear_output()
                        print(f"Has seleccionado: {selected_item.upper()}")

                def on_leaf_click_item(b):
                    global selected_item
                    selected_item = "hoja"
                    with output_message_item:
                        clear_output()
                        print(f"Has seleccionado: {selected_item.upper()}")

                def on_confirm_item_click(b):
                    clear_output()
                    final_image_to_display = None
                    message = ""

                    # Determine final image based on beverage, fruit, and item
                    if selected_beverage == "leche":
                        if selected_fruit == "fresa":
                            if selected_item == "juguete": final_image_to_display = milk_strawberry_toy_img; message = "¡Leche con fresa y juguete!"
                            elif selected_item == "hoja": final_image_to_display = milk_strawberry_leaf_img; message = "¡Leche con fresa y hoja!"
                        elif selected_fruit == "limón":
                            if selected_item == "juguete": final_image_to_display = milk_lemon_toy_img; message = "¡Leche con limón y juguete!"
                            elif selected_item == "hoja": final_image_to_display = milk_lemon_leaf_img; message = "¡Leche con limón y hoja!"
                        elif selected_fruit == "banana":
                            if selected_item == "juguete": final_image_to_display = milk_banana_toy_img; message = "¡Leche con banana y juguete!"
                            elif selected_item == "hoja": final_image_to_display = milk_banana_leaf_img; message = "¡Leche con banana y hoja!"
                    elif selected_beverage == "agua":
                        if selected_fruit == "fresa":
                            if selected_item == "juguete": final_image_to_display = water_strawberry_toy_img; message = "¡Agua con fresa y juguete!"
                            elif selected_item == "hoja": final_image_to_display = water_strawberry_leaf_img; message = "¡Agua con fresa y hoja!"
                        elif selected_fruit == "limón":
                            if selected_item == "juguete": final_image_to_display = water_lemon_toy_img; message = "¡Agua con limón y juguete!"
                            elif selected_item == "hoja": final_image_to_display = water_lemon_leaf_img; message = "¡Agua con limón y hoja!"
                        elif selected_fruit == "banana":
                            if selected_item == "juguete": final_image_to_display = water_banana_toy_img; message = "¡Agua con banana y juguete!"
                            elif selected_item == "hoja": final_image_to_display = water_banana_leaf_img; message = "¡Agua con banana y hoja!"
                    elif selected_beverage == "cerveza":
                        if selected_fruit == "fresa":
                            if selected_item == "juguete": final_image_to_display = beer_strawberry_toy_img; message = "¡Cerveza con fresa y juguete!"
                            elif selected_item == "hoja": final_image_to_display = beer_strawberry_leaf_img; message = "¡Cerveza con fresa y hoja!"
                        elif selected_fruit == "limón":
                            if selected_item == "juguete": final_image_to_display = beer_lemon_toy_img; message = "¡Cerveza con limón y juguete!"
                            elif selected_item == "hoja": final_image_to_display = beer_lemon_leaf_img; message = "¡Cerveza con limón y hoja!"
                        elif selected_fruit == "banana":
                            if selected_item == "juguete": final_image_to_display = beer_banana_toy_img; message = "¡Cerveza con banana y juguete!"
                            elif selected_item == "hoja": final_image_to_display = beer_banana_leaf_img; message = "¡Cerveza con banana y hoja!"

                    if final_image_to_display:
                        print(message)
                        display(IPython.display.Image(filename=final_image_to_display, width=700))
                    else:
                        print("No has seleccionado ningún objeto. Por favor, elige uno.")

                    next_chaos_btn = widgets.Button(description="BEBER SHAKER CHAOS", button_style='warning')
                    display(next_chaos_btn)
                    next_chaos_btn.on_click(lambda b: show_shaker_chaos_scene())

                toy_btn.on_click(on_toy_click_item)
                leaf_btn.on_click(on_leaf_click_item)
                confirm_item_btn.on_click(on_confirm_item_click)

                display(
                    widgets.VBox([
                        widgets.HBox([toy_btn, leaf_btn]),
                        confirm_item_btn,
                        output_message_item
                    ])
                )

            strawberry_btn.on_click(on_strawberry_click_fruit)
            lemon_btn.on_click(on_lemon_click_fruit)
            banana_btn.on_click(on_banana_click_fruit)
            confirm_fruit_btn.on_click(lambda b: show_item_selection_on_fruit_screen())

            display(
                widgets.VBox([
                    widgets.HBox([strawberry_btn, lemon_btn, banana_btn]),
                    confirm_fruit_btn,
                    output_message_fruit
                ])
            )


    milk_btn.on_click(on_milk_click)
    water_btn.on_click(on_water_click)
    beer_btn.on_click(on_beer_click)
    confirm_beverage_btn.on_click(lambda b: show_fruit_selection_on_beverage_screen())

    display(widgets.HBox([milk_btn, water_btn, beer_btn]), confirm_beverage_btn, output_message_beverage)
    #show_menu()
```
